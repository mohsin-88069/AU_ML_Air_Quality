### Calculating Sub Indexes of different features affecting AQI based on Correlation Matrix

```python
# PM10 Sub-Index calculation
def get_PM10_subindex(x):
    if x <= 50:
        return x
    elif x > 50 and x <= 100:
        return x
    elif x > 100 and x <= 250:
        return 100 + (x - 100) * 100 / 150
    elif x > 250 and x <= 350:
        return 200 + (x - 250)
    elif x > 350 and x <= 430:
        return 300 + (x - 350) * 100 / 80
    elif x > 430:
        return 400 + (x - 430) * 100 / 80
    else:
        return 0
    
df["PM10_SubIndex"] = df["PM10"].astype(int).apply(lambda x: get_PM10_subindex(x))
```

```python
# PM2.5 Sub-Index calculation
def get_PM25_subindex(x):
    if x <= 30:
        return x * 50 / 30
    elif x > 30 and x <= 60:
        return 50 + (x - 30) * 50 / 30
    elif x > 60 and x <= 90:
        return 100 + (x - 60) * 100 / 30
    elif x > 90 and x <= 120:
        return 200 + (x - 90) * 100 / 30
    elif x > 120 and x <= 250:
        return 300 + (x - 120) * 100 / 130
    elif x > 250:
        return 400 + (x - 250) * 100 / 130
    else:
        return 0

df["PM2.5_SubIndex"] = df["PM2.5"].astype(int).apply(lambda x: get_PM25_subindex(x))
```

```python
# SO2 Sub-Index calculation
def get_SO2_subindex(x):
    if x <= 40:
        return x * 50 / 40
    elif x > 40 and x <= 80:
        return 50 + (x - 40) * 50 / 40
    elif x > 80 and x <= 380:
        return 100 + (x - 80) * 100 / 300
    elif x > 380 and x <= 800:
        return 200 + (x - 380) * 100 / 420
    elif x > 800 and x <= 1600:
        return 300 + (x - 800) * 100 / 800
    elif x > 1600:
        return 400 + (x - 1600) * 100 / 800
    else:
        return 0

df["SO2_SubIndex"] = df["SO2"].astype(int).apply(lambda x: get_SO2_subindex(x))
```

```python
# NO2 Sub-Index calculation
def get_NO2_subindex(x):
    if x <= 40:
        return x * 50 / 40
    elif x > 40 and x <= 80:
        return 50 + (x - 40) * 50 / 40
    elif x > 80 and x <= 180:
        return 100 + (x - 80) * 100 / 100
    elif x > 180 and x <= 280:
        return 200 + (x - 180) * 100 / 100
    elif x > 280 and x <= 400:
        return 300 + (x - 280) * 100 / 120
    elif x > 400:
        return 400 + (x - 400) * 100 / 120
    else:
        return 0

df["NO2_SubIndex"] = df["NO2"].astype(int).apply(lambda x: get_NO2_subindex(x))
```

```python
# Ozone Sub-Index calculation
def get_Ozone_subindex(x):
    if x <= 50:
        return x * 50 / 50
    elif x > 50 and x <= 100:
        return 50 + (x - 50) * 50 / 50
    elif x > 100 and x <= 168:
        return 100 + (x - 100) * 100 / 68
    elif x > 168 and x <= 208:
        return 200 + (x - 168) * 100 / 40
    elif x > 208 and x <= 748:
        return 300 + (x - 208) * 100 / 539
    elif x > 748:
        return 400 + (x - 400) * 100 / 539
    else:
        return 0

df["Ozone_SubIndex"] = df["Ozone"].astype(int).apply(lambda x: get_Ozone_subindex(x))
```

```python
# Removing columns
df.drop(['Date', 'Time', 'Temp', 'RH'], axis=1, inplace=True)
```

```python
def get_CO_subindex(x):
    if x <= 1:
        return x * 50 / 1
    elif x <= 2:
        return 50 + (x - 1) * 50 / 1
    elif x <= 10:
        return 100 + (x - 2) * 100 / 8
    elif x <= 17:
        return 200 + (x - 10) * 100 / 7
    elif x <= 34:
        return 300 + (x - 17) * 100 / 17
    elif x > 34:
        return 400 + (x - 34) * 100 / 17
    else:
        return 0

df["CO_SubIndex"] = df["CO"].apply(lambda x: get_CO_subindex(x))
```

```python
df["AQI"] = (round(df[['PM10_SubIndex', 'PM2.5_SubIndex', 'SO2_SubIndex', 'NO2_SubIndex', 'Ozone_SubIndex']].max(axis = 1)))
```

```python
df.columns
```

```python
def AQI_Range(x):
    if x<=50:
        return "Good"
    elif x>50 and x<=100:
        return "Moderate"
    elif x>100 and x<=200:
        return "Poor"
    elif x>200 and x<=300:
        return "Unhealthy"
    elif x>300 and x<=400:
        return "Very unhealthy"
    elif x>400:
        return "Hazardous"

df['AQI_Range'] = df['AQI'] .apply(AQI_Range)
```

```python
def get_AQI_Bucket(x):
    if x=='Good':
        return 0
    elif x=='Moderate':
        return 1
    elif x=='Poor':
        return 2
    elif x=='Unhealthy':
        return 3
    elif x=='Very unhealthy':
        return 4
    else:
        return 5
    
df["AQI_Bucket"]=df["AQI"].apply(lambda x: get_AQI_Bucket(x))
```

## Applying Unsupervised Algorithm

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
X=pd.DataFrame(X)
X.head()
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
X_pca=pca.fit_transform(X)
```

```python
X_pca=pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
X_pca
```

### K Means Clustering

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=2)
y_predicted = km.fit_predict(X_pca[['PC1','PC2']])
y_predicted
```

```python
X_pca['KM_2_cluster']=y_predicted
X_pca.sample(10)
```

```python
km.cluster_centers_
```

```python
X_pca0 = X_pca[X_pca.KM_2_cluster==0]
X_pca1 = X_pca[X_pca.KM_2_cluster==1]
# X_pca2 = X_pca[X_pca.KM_cluster==2]
plt.scatter(X_pca0.PC1,X_pca0.PC2,color='green')
plt.scatter(X_pca1.PC1,X_pca1.PC2,color='red')
# plt.scatter(X_pca2.PC1,X_pca2.PC2,color='blue')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.legend()
```

```python
km.cluster_centers_
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(X_pca[['PC1','PC2']])
    sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```

### Optimal Value of K is 3

```python
km_opt = KMeans(n_clusters=3)
y_predicted = km_opt.fit_predict(X_pca[['PC1','PC2']])
y_predicted
```

```python
X_pca['KM_3_cluster']=y_predicted
X_pca.sample(10)
```

```python
km_opt.cluster_centers_
```

```python
X_pca0_opt = X_pca[X_pca.KM_3_cluster==0]
X_pca1_opt = X_pca[X_pca.KM_3_cluster==1]
X_pca2_opt = X_pca[X_pca.KM_3_cluster==2]
plt.scatter(X_pca0_opt.PC1,X_pca0_opt.PC2,color='green')
plt.scatter(X_pca1_opt.PC1,X_pca1_opt.PC2,color='red')
plt.scatter(X_pca2_opt.PC1,X_pca2_opt.PC2,color='blue')
plt.scatter(km_opt.cluster_centers_[:,0],km_opt.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.legend()
```

## DBSCAN Clustering

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
db = DBSCAN(eps=1, min_samples=3)
y_predicted = db.fit(X_pca)
```

```python
s=pd.DataFrame(db.labels_)
s.value_counts()
# -1 is noise point (outliers)
# 4 clusters are being formed
```

```python
clusters=db.labels_
silhouette_score(X_pca, clusters)
#silhouette score of 0.5 indicates a good cluster formation with minority overlapping
```
